## O pacote numpy
https://numpy.org/

Este pacote serve para operações matriciais e estatísticas básicas

```python
# Importando o numpy
import numpy as np
```

```python
# Criando um número aleatório com numpy

c = np.random.rand(100)
#print(c)
print(type(c))

# Acessando o último valor do array numpy
print('O último valor de c é: ' + str(c[-1]))
print('\n')

# Acessando os 10 primeiros valores
print('Os 10 primeiros valores da lista são: ' + str(c[0:10]))
print('\n')

# Loop com um numpy array
for ii in range(0,10):
    print(str(ii) + ' ' + str(c[ii]))

print('\n')

# O comrpimento do numpy array - shape da matriz
print('O shape da matriz a é: ' + str(np.shape(c)))
```

```python
d = np.random.rand(100,100)
print(d)

print('Qual é o shape/dimensão da matriz?')
print(np.shape(d))

# Linha 11 e coluna 11
print('Indexação numpy-matriz 2D')
print(d[10,10])
```

```python
# Gerando uma matriz 3D

mat3D = np.random.rand(365,100,100)
print(np.shape(mat3D))
```

```python
# Acessando todos os valores da primeira dimensão na célula 49, 49

# Todos os valores da série
timeSeries = mat3D[:,49,49]
print(timeSeries)

# 10 primeiros valores da série
timeSeries = mat3D[:10,49,49]
print(timeSeries)

# 10 últimos valores da série
timeSeries = mat3D[-10:,49,49]
print(timeSeries)
```

```python
# Operações matriciais com numpy

timeSeries2 = mat3D[:,50:52,50]
print(np.shape(timeSeries2))
print(timeSeries2.shape)

# Matriz transposta
timeSeriesTransposta = timeSeries2.transpose()
print(timeSeriesTransposta.shape)
```

```python
# Estatísticas com numpy

# Média de todos os valores
print('Essa é a média de todos os valores')
mediaTS = timeSeries2.mean()
print(mediaTS)

# Média nas colunas
mediaTScolumns = np.mean(timeSeries2,axis=0)
print('Essa é a média de cada coluna')
print(mediaTScolumns)

# Média nas linhas
mediaTSrows = np.mean(timeSeries2,axis=1)
print('Essa é a média de cada linha')
print(mediaTSrows)
```

```python
# Média, mínimo, máximo, desvio, percentil, mediana, variancia

# Média
media3D = np.mean(mat3D)
print(media3D)

# Mínimo
min3D = np.min(mat3D)
print(min3D)

# Máximo
max3D = np.max(mat3D)
print(max3D)

# Desvio
desv3D = np.std(mat3D)
print(desv3D)

# Percentil
percentil3D = np.percentile(mat3D,90)
print(percentil3D)

# Mediana
mediana3D = np.median(mat3D)
print(mediana3D)

# Variância
var3D = np.var(mat3D)
print(var3D)
```

```python
# Concatenação
mat3D2 = mat3D.copy()

matConcat = np.concatenate([mat3D, mat3D2])
print(matConcat.shape)
```

```python
# Concatenação em lista

listMat = [mat3D, mat3D, mat3D]

mat3Dconc = np.concatenate(listMat)
print(mat3Dconc.shape)
```

## Plotando com o matplotlib

https://matplotlib.org/

* plot
* histograma
* bar
* boxplot
* pcolor
* scatter

```python
# Importando o pacot matplotlib
import matplotlib.pyplot as plt

# Criando um objeto matplotlib para plotar as figuras
fig,ax = plt.subplots(3, figsize= (15,10))

# lineplot
ax[0].plot(mat3D[:,10,10])
ax[1].plot(mat3D[:,10:12,10], alpha=0.5)

# bar
ax[2].bar(np.array(range(0,mat3D.shape[0])),mat3D[:,10,10])
print(type(range(0,mat3D.shape[0])))
```

```python
# BOXPLOT

fig, ax = plt.subplots(2)
ax[0].boxplot(mat3D[:,10:20,0].flatten())
ax[1].boxplot(mat3D[:,10:20,0])
```

```python

# pcolor - gráfico de superfície

fig, ax = plt.subplots(2,2)

ax[0,0].pcolor(mat3D[0,:,:],cmap='gray')
ax[0,1].contour(np.sin(mat3D[0,:,:]))
ax[1,1].scatter(np.sin(mat3D[0,:,:]).flatten(),mat3D[0,:,:].flatten())
ax[1,0].hist(np.sin(mat3D[0,:,:]).flatten())
```
